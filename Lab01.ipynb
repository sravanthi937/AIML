##generating scatterplot(part2)

```python
people=['A','B','C','D','E','F','G','H','I','J','K','L']
age=[21,12,32,45,37,18,52,5,40,48,15]
weight=[55,35,77,68,70,60,72,69,18,65,82,48]
height=[160,135,170,165,173,168,175,159,105,171,155,158]
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of patients in kgs")
plt.show
```

```python
people=['A','B','C','D','E','F','G','H','I','J','K','L']
age=[21,12,32,45,37,18,52,5,40,48,15]
weight=[55,35,77,68,70,60,72,69,18,65,82,48]
height=[160,135,170,165,173,168,175,159,105,171,155,158]
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, width=bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in index], weight, width=bar_width, color='red', label='Weight')
plt.title('Bar Chart - Health Care Data')
plt.xlabel('People')
plt.ylabel('Height and Weight')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.grid(True)
plt.show()
```

```python
people=['A','B','C','D','E','F','G','H','I','J','K','L']
age=[21,12,32,45,37,18,52,5,40,48,15]
weight=[55,35,77,68,70,60,72,69,18,65,82,48]
height=[160,135,170,165,173,168,175,159,105,171,155,158]
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.hist(height, bins=10, color='blue', alpha=0.7, label='Height')
plt.hist(weight, bins=10, color='red', alpha=0.7, label='Weight')
plt.title('Histogram - Health Care Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
